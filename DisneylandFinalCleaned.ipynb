# Creating a function to convert the 'Established' column into a format of only 'Year' values

```python
def to_year(val):
    val = str(val).strip()
    if val.upper() == 'TBA':
        return 'TBA'
    try:
        year = pd.to_datetime(val, errors='coerce').year
        return str(int(year)) if pd.notnull(year) else 'TBA'
    except:
        return 'TBA'

disney['Established'] = disney['Established'].apply(to_year)
```
