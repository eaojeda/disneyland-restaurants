# Importing necessary libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Creating a variable for the URL

```python
url = "https://allears.net/dlr/dining/location/"
```

# Setting up a variable to request data from URL and verifying it works correctly

```python
response = requests.get(url)
response.status_code
```

# Extracting html data with BeautifulSoup

```python
soup = BeautifulSoup(response.content, 'html')
```

# Loading entire html with soup variable

```python
soup
```

# Extracting and finding only 'div' within soup

```python
soup.find_all('div')
```

# Finding all divs within the container class and creating a variable named header

```python
header = soup.find_all('div', {'class': 'container'})
```

# Finding all h2 within header and creating a variable named theme_park

```python
theme_park = [head.find_all('h2') for head in header]
```

```python
theme_park
```

# Stripping text files within themepark to return only names

```python
theme_park_names = [tag.get_text(strip=True) for sublist in theme_park for tag in sublist]
```

```python
theme_park_names
```

# Finding all divs within the facetedsearch-results class and creating header2

```python
header2 = soup.find_all('div', {'class': 'facetedsearch-results'})
```

# Finding all h3 within the header2 variable

```python
restuarants = [head.find_all('h3') for head in header2]
```

```python
restuarants
```

# Stripping text from restaurants variable to return only restaurant locations

```python
restuarant_locations = [tag.get_text(strip=True) for sublist in restuarants for tag in sublist]
```

```python
restuarant_locations
```

# Finding all divs within the 'fsearch_subrendergroupitems' class and assigning it to header3

```python
header3 = soup.find_all('div', {'class': 'fsearch_subrendergroupitems'})
```

```python
header3
```

# Finding all 'a' tags within the header3 variable and assigning it to a new variable named restaurant_name

```python
restaurant_name = [head.find_all('a') for head in header3]
```

```python
restaurant_name
```

# Stripping only text values from the restaurant_name variable

```python
restaurant_names = [tag.get_text(strip=True) for sublist in restaurant_name for tag in sublist] 
```

```python
restaurant_names
```

# Creating column names for a dataframe

```python
columns = ['Theme Park', 'Location', 'Restaurant']
```

# Creating a dataframe

```python
df = pd.DataFrame(columns=columns)
```
